```python
# df["VALUE"] column has a negative rainfall which does't make any sense So, we make those value to zero

column_to_replace = 'VALUE'

# Loop to replace negative values with 0 in the specified column
for index, row in df.iterrows():
    if row[column_to_replace] < 0:
        df.at[index, column_to_replace] = 0
```

```python
df.head(100)
```

```python
# Dropna a usedless columns like df["Unnamed: 0"] , df['FREQUENCY'] , df['VARIABLE_NAME'] and df["VARIABLE_NOTES"]
column_to_drop = ['Unnamed: 0','FREQUENCY','VARIABLE_NAME','VARIABLE_NOTES']
df = df.drop(columns=column_to_drop)
```
